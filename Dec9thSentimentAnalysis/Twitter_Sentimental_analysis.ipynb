```
import nltk
nltk.download("vader_lexicon")
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```
import pandas as pd
```

```
twitter = pd.read_csv("/content/twitter_data.csv",encoding="latin") #latin to prevent encoding error
twitter.head(20)
```

```
twitter.iloc[0]['SentimentText']
```

```
twitter.iloc[0,2]
```

```
sia.polarity_scores(twitter.iloc[0]['SentimentText'])
```

```
len(twitter)
```

```
polarity=[]
for i in range(len(twitter)):
  polarity.append(sia.polarity_scores(twitter.iloc[i]['SentimentText']))
```

```
polarity
```

```
twitter['Polarity Scores'] = polarity
twitter.head()
```

```
twitter.iloc[0]["Polarity Scores"]['compound']
```

```
compound =[]
for i in range(len(twitter)):
  compound.append(twitter.iloc[i]["Polarity Scores"]['compound'])
```

```
twitter['compound'] = compound
```

```
twitter.head()
```

```
twitter['Sentiment'].value_counts()
```

```
sentiment = []
for i in range(len(twitter)):
  c = twitter.iloc[i]['compound']
  if c > 0:
    sentiment.append(1)
  else:
    sentiment.append(0)
```

```
twitter['Predicted Label'] = sentiment
twitter.head(20)
```

```
count = 0
for i in range(len(twitter)):
  Senti = twitter.iloc[i]['Sentiment']
  pred_label = twitter.iloc[i]['Predicted Label']
  if Senti == pred_label:
    count = count+1
```

```
count
```

```
accuracy = count/len(twitter)*100
accuracy
```

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(twitter['Sentiment'],twitter['Predicted Label'])
cm
```

```
from sklearn.metrics import accuracy_score
a = accuracy_score(twitter['Sentiment'],twitter['Predicted Label'])
a*100
```
